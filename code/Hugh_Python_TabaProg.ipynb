```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly.express as px
pd.set_option("display.max_columns", None)
pd.options.display.float_format = '{:,.2f}'.format
```

```python
def combine_csv_files(file_paths):
    dataframes = []
    for file in file_paths:
        try:
            df = pd.read_csv(file)
            dataframes.append(df)
        except Exception as e:
            print(f"Error reading {file}: {e}")
    combined_df = pd.concat(dataframes, ignore_index=True)
    return combined_df
```

```python
csv_files = ["united states 2016-01-01 to 2016-12-31.csv",
             "united states 2017-01-01 to 2017-12-31.csv",
             "united states 2018-01-01 to 2018-12-31.csv",
             "united states 2019-01-01 to 2019-12-31.csv",
             "united states 2020-01-01 to 2020-12-31.csv",
             "united states 2021-01-01 to 2021-12-31.csv", 
             "united states 2022-01-01 to 2022-12-31.csv"]
combined_data = combine_csv_files(csv_files)
```

```python
# Specify the columns to visualize
selected_columns = ["avg_max_temp",	"avg_min_temp","avg_temp","avg_feels_like_max_temp",	"avg_feels_like_min_temp",
                    "avg_feels_like_temp",	"avg_dew_point",	
                    "avg_humidity",	"total_precipitation","avg_precipitation_probability","avg_precipitation_cover",	"total_snow",	
                    "avg_snow_depth","avg_wind_gust","avg_wind_speed","avg_wind_direction",	"avg_sea_level_pressure",	
                    "avg_cloud_cover",	"avg_visibility",	"avg_solar_radiation",	"total_solar_energy",	"avg_uv_index",	"avg_moon_phase",	
                    "average_cereal_production_per_week"
]

# Create a figure for the subplots
n_cols = 5
n_rows = (len(selected_columns) + n_cols - 1) // n_cols  # Calculate number of rows needed
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 4 * n_rows))
axes = axes.flatten()  # Flatten the 2D array of axes to 1D for easy indexing

# Get the SET2 color palette
set2_palette = sns.color_palette("Set2", n_colors=len(selected_columns))

# Loop through the selected columns and create plots
for i, col in enumerate(selected_columns):
    sns.histplot(df[col], ax=axes[i], bins=30, kde=True, color=set2_palette[i])
    axes[i].set_title(col)
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Frequency")

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Add the title 
plt.suptitle("Distribution of Weather Data columns", fontsize=16, fontweight='bold')

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.95]) 
plt.show()
```

```python
categorical_columns = ["most_common_conditions","most_common_icon"]
num_rows = (len(categorical_columns) + 2) // 3

# Create subplots
fig, axes = plt.subplots(num_rows, 3, figsize=(15, 5 * num_rows))
axes = axes.flatten()  # Flatten to easily iterate

# Create a bar plot for each categorical column
for idx, column in enumerate(categorical_columns):
    # Generate the bar plot, assigning x to hue and disabling legend
    sns.countplot(data=df, x=column, hue=column, ax=axes[idx], palette='Set2', legend=False)
    
    #Add mean line
    counts = df[column].value_counts()
    mean_count = counts.mean()
    axes[idx].axhline(mean_count, color='red', linestyle='--', label='Mean')
    
    #Add labels and title
    axes[idx].set_title(f'Count of {column}')
    axes[idx].set_xlabel(column)
    axes[idx].set_ylabel("Count")
    axes[idx].tick_params(axis="x", rotation=45)  # Rotate x labels for better visibility
    
    # Add legend for the mean line
    axes[idx].legend()

# Hide any unused axes (if any)
for j in range(idx + 1, len(axes)):
    fig.delaxes(axes[j])
    
plt.suptitle("Distribution of Categorical Variables", fontsize=16, fontweight='bold', x=0.35)


plt.tight_layout(rect=[0, 0, 1, 0.95]) 
plt.show()
```

# Machine Learning Methods
## Random Forest Regression

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
label_encoder = LabelEncoder()
df["most_common_conditions"] = label_encoder.fit_transform(df["most_common_conditions"])
df["most_common_icon"] = label_encoder.fit_transform(df["most_common_icon"])
df["most_common_description"] = label_encoder.fit_transform(df["most_common_description"])
df["most_common_station"] = label_encoder.fit_transform(df["most_common_station"])
df['year'] = df["week_starting_date"].dt.year
df['month'] = df["week_starting_date"].dt.month
df['day_starting'] = df["week_starting_date"].dt.day
df = df.drop(columns=["week_starting_date"])
df.head()
```

```python
X = df.drop(columns =["avg_feels_like_max_temp",	"avg_feels_like_min_temp",	
                      "avg_feels_like_temp","average_cereal_production_per_week"], 
                      axis=1)
y_reg = df['average_cereal_production_per_week']
#TTTS for regression
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X, y_reg, test_size=0.3, random_state=42)
```

```python
#Train the Random Forest Regressor
regressor = RandomForestRegressor(random_state=42)
regressor.fit(X_train_reg, y_train_reg)
#Predict and Evaluate
y_pred_reg = regressor.predict(X_test_reg)
```

```python
###Regression problem evaluation
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
#Mean Squared Error
mse = mean_squared_error(y_test_reg, y_pred_reg)
print(f"Mean Squared Error (MSE): {mse:.2f}")
#Root Mean Squared Error
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
#Mean Absolute Error
mae = mean_absolute_error(y_test_reg, y_pred_reg)
print(f"Mean Absolute Error (MAE): {mae:.2f}")
#R^2 Score
r2 = r2_score(y_test_reg, y_pred_reg)
print(f"R (Coefficient of Determination): {r2:.2f}")
```

## Support Vector Regression

```python
from sklearn.svm import SVR
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X, y_reg, test_size=0.3, random_state=42)
model = SVR(kernel='rbf', C=1, epsilon=0.1)
model.fit(X_train_reg, y_train_reg)
y_pred = model.predict(X_test_reg)
```

```python
mae = mean_absolute_error(y_test_reg, y_pred)
mse = mean_squared_error(y_test_reg, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test_reg, y_pred)

# Print the evaluation metrics
print("Evaluation Metrics for SVR:")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R Score: {r2:.4f}")
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X, y_reg, test_size=0.3, random_state=42)
model = LinearRegression()
model.fit(X_train_reg, y_train_reg)
y_pred = model.predict(X_test_reg)
```

```python
mae = mean_absolute_error(y_test_reg, y_pred)
mse = mean_squared_error(y_test_reg, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test_reg, y_pred)

print("Coefficients (slope):", model.coef_)
print("Intercept:", model.intercept_)
print("---")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R Score: {r2:.4f}")
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X, y_reg, test_size=0.8, random_state=42)
model = DecisionTreeRegressor(criterion='squared_error', max_depth=4, random_state=42)
model.fit(X_train_reg, y_train_reg)
y_pred = model.predict(X_test_reg)
```

```python
mae = mean_absolute_error(y_test_reg, y_pred)
mse = mean_squared_error(y_test_reg, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test_reg, y_pred)

print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R Score: {r2:.4f}")
```

## Gradient Boosting Regression

```python
from sklearn.ensemble import GradientBoostingRegressor
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X, y_reg, test_size=0.3, random_state=42)
model = GradientBoostingRegressor(n_estimators=50,learning_rate=0.2,random_state=42)
model.fit(X_train_reg, y_train_reg)
y_pred = model.predict(X_test_reg)
```

```python
mae = mean_absolute_error(y_test_reg, y_pred)
mse = mean_squared_error(y_test_reg, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test_reg, y_pred)

print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R Score: {r2:.4f}")
```
